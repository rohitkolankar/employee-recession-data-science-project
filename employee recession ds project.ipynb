# data preprocessing

```python
salary_new = pd.get_dummies(main_df['salary'])
```

```python
salary_new
```

```python
department_new = pd.get_dummies(main_df['department'])
```

```python
department_new
```

```python
new_df = pd.concat([main_df,salary_new,department_new],axis=1)
```

```python
new_df
```

```python
new_df.drop(['salary','department'],axis=1,inplace=True)
```

```python
new_df.info()
```

```python
X = new_df.drop("left",axis=1)
y = new_df['left']
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# data training and predicting

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
DT = DecisionTreeClassifier()
DT.fit(X_train,y_train)
```

```python
score = DT.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
dt_score = accuracy_score(y_test,score)
```

```python
dt_score
```

```python
RF = RandomForestClassifier()
RF.fit(X_train,y_train)
score2 = RF.predict(X_test)
```

```python
rf_score = accuracy_score(y_test,score2)
```

```python
rf_score
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler().fit(X_train)
X_train_std=sc.transform(X_train)
X_test_std=sc.transform(X_test)
```

```python
KNN  =  KNeighborsClassifier()
KNN.fit(X_train_std,y_train)
score3 = KNN.predict(X_test_std)
```

```python
knn_score = accuracy_score(y_test,score3)
knn_score
```

```python
k_range = range(1,30)
scores={}
scores_list=[]
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    score  = knn.predict(X_test)
    
    scores[k]  = accuracy_score(score,y_test)*100
    scores_list.append( accuracy_score(score,y_test)*100)
```
